# What is NumPy?  
NumPy (Numerical Python) is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.

merge() enables SQL style join types along specific columns. See the Database style joining section.

### Grouping

```python
df = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)
df
```

Grouping and then applying the sum() function to the resulting groups

```python
df.groupby("A")[["C", "D"]].sum()
```

Grouping by multiple columns forms a hierarchical index, and again we can apply the sum() function

```python
df.groupby(["A", "B"]).sum()
```

# Dados Reais

```python
import json
import pandas as pd
import requests
```

```python
def recuperarDados(estadoEstacao):
  response = requests.get(f'https://apitempo.inmet.gov.br/estacao/diaria/2020-01-01/2020-12-30/{estadoEstacao}')
  data = json.loads(response.text)
  df = pd.json_normalize(data)
  return df
```

```python
recife = recuperarDados("A301")
recife['DT_MEDICAO'] = pd.to_datetime(recife['DT_MEDICAO'])
recife.set_index('DT_MEDICAO', inplace=True)
recife.head()
```

```python
recife['CHUVA'][0:5].sum()
```

```python
recife.info()
```

```python
recife['CHUVA'][0:5].astype('float').sum()
```

```python
recife['CHUVA'].dtypes
```

```python
df_teste = recife
for column in df_teste.columns:
  if df_teste[column].dtypes == 'O' and column != 'DC_NOME' and column != 'UF' and column != 'CD_ESTACAO':
    df_teste[column] = df_teste[column].astype('float')

df_teste.info()
```

```python
chuva_semanal = df_teste['CHUVA'].resample('W').sum()
chuva_semanal.head()
```

# PLOTLY

The plotly.express module (usually imported as px) contains functions that can create entire figures at once, and is referred to as Plotly Express or PX.

[Referncia](https://plotly.com/python/plotly-express/)

```python
import plotly.express as px
```

```python
df = px.data.iris()
fig = px.scatter_matrix(df, dimensions=["sepal_width", "sepal_length", "petal_width", "petal_length"], color="species")
fig.show()
```

```python

df = px.data.gapminder()

fig = px.scatter(df.query("year==2007"), x="gdpPercap", y="lifeExp",
	         size="pop", color="continent",
                 hover_name="country", log_x=True, size_max=60)
fig.show()
```

```python
df = px.data.gapminder().query("year == 2007")
fig = px.treemap(df, path=[px.Constant('world'), 'continent', 'country'], values='pop',
                  color='lifeExp', hover_data=['iso_alpha'])
fig.show()
```

```python
df = px.data.tips()
fig = px.histogram(df, x="total_bill", y="tip", color="sex", marginal="rug", hover_data=df.columns)
fig.show()
```

```python
df = px.data.tips()
fig = px.violin(df, y="tip", x="smoker", color="sex", box=True, points="all", hover_data=df.columns)
fig.show()
```

```python
df = px.data.iris()
fig = px.density_heatmap(df, x="sepal_width", y="sepal_length", marginal_x="rug", marginal_y="histogram")
fig.show()
```

```python
df = px.data.wind()
fig = px.bar_polar(df, r="frequency", theta="direction", color="strength", template="plotly_dark",
            color_discrete_sequence= px.colors.sequential.Plasma_r)
fig.show()
```

```python
df = px.data.carshare()
fig = px.scatter_mapbox(df, lat="centroid_lat", lon="centroid_lon", color="peak_hour", size="car_hours",
                  color_continuous_scale=px.colors.cyclical.IceFire, size_max=15, zoom=10,
                  mapbox_style="carto-positron")
fig.show()
```
